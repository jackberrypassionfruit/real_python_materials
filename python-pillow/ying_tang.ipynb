```python
from PIL import Image, ImageFilter
filename = 'dot_and_hole.jpg'
with Image.open(filename) as img:
    img.load()
    
img = img.resize((img.width*2, img.height*2))

def erode(erosions, img):
    for _ in range(erosions):
        img = img.filter(ImageFilter.MinFilter(3))
    return img


def dilate(dilations, img):
    for _ in range(dilations):
        img = img.filter(ImageFilter.MaxFilter(3))
    return img
    
```

```python
num_iterations = 15
# img_eroded = img
# for _ in range(num_iterations):
#     img_eroded = img_eroded.filter(ImageFilter.MinFilter(3))
img_eroded = erode(img, erosions=num_iterations)
display(img_eroded)

# img_dilated = img_eroded
# for _ in range(num_iterations):
#     img_dilated = img_dilated.filter(ImageFilter.MaxFilter(3))
img_dilated = dilate(img_eroded, dilations=num_iterations)
display(img_dilated)
```
